```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/tmdb_5000_movies.csv")
df.head()
```

```python
df.info()
```

```python
# Keep required columns

df = df[['title', 'genres', 'overview']]
df.head()
```

```python
#Handling missing value

df['overview'] = df['overview'].fillna('')
df['genres'] = df['genres'].fillna('')
```

```python
# combine features of genres and overview

df['combined_features'] = df['genres'] +" " +df['overview']
df[['title', 'combined_features']].head()
```

```python
#convert movie text into numeric value

from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words = 'english')

tfidf_matrix = tfidf.fit_transform(df['combined_features'])

tfidf_matrix.shape
```

```python
# here each value will represent how similar two movies are

from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

cosine_sim.shape
```

```python
indices = pd.Series(df.index, index=df['title']).drop_duplicates()

def recommend_movies(movie_title, num_recommendations=5):
    if movie_title not in indices:
        return "Movie not found "
    
    idx = indices[movie_title]

    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    sim_scores = sim_scores[1:num_recommendations+1]

    movie_indices = [i[0] for i in sim_scores]

    return df['title'].iloc[movie_indices]
```

```python
recommend_movies("Inception")
```
